# 项目：分析鸢尾花种类数据

## 分析目标

此数据分析报告的目的是基于鸢尾花的属性数据，分析两种鸢尾花萼片、花瓣的长度和宽度平均值，是否存在显著性差异，让我们可以对不同种类鸢尾花的属性特征进行推断。

## 简介

原始数据`Iris.csv`包括两种鸢尾花，每种有 50 个样本，以及每个样本的一些属性，包括萼片的长度和宽度、花瓣的长度和宽度。

`Iris.csv`每列的含义如下：
- Id：样本的ID。
- SepalLengthCm：萼片的长度（单位为厘米）。
- SepalWidthCm：萼片的宽度（单位为厘米）。
- PetalLengthCm：花瓣的长度（单位为厘米）。
- PetalWidthCm：花瓣的宽度（单位为厘米）。
- Species：鸢尾花种类。

## 探索数据

在开始数据分析前，我们可以先借助数据可视化，探索`Setosa`和`Versicolor`这两种鸢尾花的变量特点。

数据可视化可以让我们对数据有个更直观的了解，比如了解数据的分布、发现变量之间的关系等等，为后续分析提供方向。

同时，由于此次分析目的是了解不同种类鸢尾花的属性特征是否存在差异，我们可以利用颜色对两种鸢尾花进行分类。

```python
sns.pairplot(cleaned_data,hue='Species')
plt.show()
```

从输出结果可知，`Setosa`和`Versicolor`样本的花瓣长度以及花瓣宽度的分布存在数据上明显的不同，可以猜测假设检验的结果时，两种鸢尾花的花瓣长度与宽度有统计显著性差异。

萼片的长度与宽度在分布上存在重叠，暂时无法通过图标下结论，需要进行假设检验，来推断总体的萼片长度和宽度之间是否有差异。

## 分析数据

我们将利用假设检验，依次检验`Setosa`和`Versicolor`这两种鸢尾花在萼片、花瓣的长度和宽度平均值方面是否存在统计显著性差异。

由于两组样本数据均为50，样本数量不够大，因此进行t检验，而不是z检验。假设此数据集样本符合t检验的两个前提：样本为随机抽样，总体呈正态分布。

先引入t检测所需要的模块。

```python
from scipy.stats import ttest_ind
```

## 分析萼片长度

`Setosa`和`Versicolor`的萼片长度分布如下：

```python
#自行设定区间开头
bins=np.arange(4.2,7.1,0.1)
```

```python

sns.histplot(data=iris_setosa['SepalLengthCm'],bins=bins)
sns.histplot(data=iris_versicolor['SepalLengthCm'],bins=bins)
plt.show()
```

#### 建立假设

  
  
$H_0$:Setosa鸢尾花和Versicolor鸢尾花萼片长度的平均值不存在在显著区别。

$H_1$:Setosa鸢尾花和Versicolor鸢尾花萼片长度的平均值存在在显著区别。



    

  

#### 确认检验是单位还是双尾


由于我们只检测平均值是否存在差异，不在乎哪个品种的萼片更长，所以是双尾检验。


#### 确定显著水平


选择0.05作为显著水平。


#### 计算t值和p值

```python
t_stat,p_value=ttest_ind(iris_setosa['SepalLengthCm'],iris_versicolor['SepalLengthCm'])
print(f't值：{t_stat}')
print(f'p值：{p_value}')
```

#### 结论

由于p值小于显著水平0.05，因此我们拒绝原假设，说明Setosa和Versicolor鸢尾花萼片长度的平均值存在显著区别。

### 分析萼片宽度

Setosa 和 Versicolor 萼片长度的分布如下。

```python
bins=np.arange(1.9,4.5,0.1)
```

```python
sns.histplot(iris_setosa['SepalWidthCm'],bins=bins)
sns.histplot(iris_versicolor['SepalWidthCm'],bins=bins)
plt.show()
```

#### 建立假设

$H_0$：Setosa鸢尾花和Versicolor鸢尾花萼片宽度的平均值不存在显著区别。


$H_1$：Setosa鸢尾花和Versicolor鸢尾花萼片宽度的平均值存在显著区别。

#### 确定检验是单尾还是双尾

由于我们只检验平均值是否存在差异，不在乎哪个品种的萼片更宽，所以是双尾检验。

#### 确定显著水平

我们将0.05作为显著水平

#### 计算t值和p值

```python
t_stat,p_value=ttest_ind(iris_setosa["SepalWidthCm"], iris_versicolor["SepalWidthCm"])
print(f"t值：{t_stat}")
print(f"p值：{p_value}")
```

#### 结论

由于p值小于显著水平0.05，我们因此拒绝原假设，说明Setosa鸢尾花和Versicolor鸢尾花萼片宽度的平均值存在显著区别。
